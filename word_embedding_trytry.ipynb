```python
train_loss_list = []
keys = list(Dic_of_T.keys())
for i in range(iters_num):
    x_data = []
    t_data = []
    
    num = random.choice(keys)
    x = wi.inform['voc_vectors'][num]
        
    for value in Dic_of_T[num]:
        x_data.append(x)    
        t_data.append(wi.inform['voc_vectors'][value])
            
    x_data_np = np.array(x_data)
    t_data_np = np.array(t_data)
    

    grad = net.gradient(x_data_np,t_data_np)
    
    #for key in ('W1','W2'):
    for key in ('W1','b1','W2','b2'):
    #for key in ('W1','b1','W2'):
        net.params[key] -= learning_rate * grad[key]
        
        loss = net.loss(x_data_np,t_data_np)
        if i % 100 == 0:
            train_loss_list.append(loss)
    
```
